```python
import pandas as pd

df = pd.read_csv('data/BTC-USD.csv', parse_dates=['Date'])

# Affichage des premières lignes pour vérification
print(df.head())

# Vérification de la structure
print(df.info())
```

```python
# Statistiques descriptives globales
stats = df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']].describe()
print(stats)
```

```python
import matplotlib.pyplot as plt

# Tracer l'évolution des prix de clôture
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['Close'], label='Close', linewidth=1)
plt.title('Évolution du BTCUSD (2014-2024)')
plt.xlabel('Date')
plt.ylabel('Prix de clôture')
plt.legend()
plt.show()

# Volume au fil du temps
plt.figure(figsize=(14, 4))
plt.plot(df['Date'], df['Volume'], color='orange', label='Volume')
plt.title('Volume quotidien BTCUSD')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()
```

```python
# Calcul de la volatilité quotidienne (High-Low)
df['Volatilité'] = df['High'] - df['Low']
volatilite_stats = df['Volatilité'].describe()
print(volatilite_stats)

# Visualisation de la volatilité
plt.figure(figsize=(14, 4))
plt.plot(df['Date'], df['Volatilité'], color='red', alpha=0.5)
plt.title('Volatilité quotidienne (High-Low)')
plt.xlabel('Date')
plt.ylabel('Volatilité')
plt.show()
```

```python
# Découpe temporelle
early = df[df['Date'] < '2017-01-01']
mid = df[(df['Date'] >= '2017-01-01') & (df['Date'] < '2020-01-01')]
late = df[df['Date'] >= '2020-01-01']

print('Volume moyen Early:', early['Volume'].mean())
print('Volume moyen Mid:', mid['Volume'].mean())
print('Volume moyen Late:', late['Volume'].mean())
```

```python
df['Ratio_O_C'] = df['Close'] / df['Open']
ratio_stats = df['Ratio_O_C'].describe()
print(ratio_stats)
```

# Pipeline

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

plt.figure(figsize=(14, 10))
ax = plt.gca()
ax.axis('off')

# Configuration des boîtes
boxes = [
    {"xy": (0.1, 0.85), "label": "Données brutes\n", "color": "#FFD700"},
    {"xy": (0.1, 0.65), "label": "Prétraitements:\n- Log-return\n- Rolling Std\n- GARCH\n- Hurst Exponent", "color": "#87CEEB"},
    {"xy": (0.1, 0.45), "label": "Normalisation\n(RobustScaler)", "color": "#98FB98"},
    {"xy": (0.1, 0.25), "label": "Séquences temporelles\n(Fenêtre: 90 jours)", "color": "#FFB6C1"},
    {"xy": (0.1, 0.05), "label": "Modèle Deep Learning\n(Conv1D + BiLSTM + Dense)", "color": "#DDA0DD"},
    {"xy": (0.4, 0.05), "label": "Prédiction\nProbabilité de pic\nde volatilité", "color": "#FFA07A"},
    {"xy": (0.7, 0.05), "label": "Alerte\n(Threshold > 0.5)", "color": "#F08080"}
]

# Dessiner les boîtes
for box in boxes:
    ax.add_patch(plt.Rectangle(
        box["xy"], 0.25, 0.12, 
        fill=True, 
        color=box["color"], 
        alpha=0.8,
        edgecolor='black',
        linewidth=1.5
    ))
    plt.text(
        box["xy"][0] + 0.125, 
        box["xy"][1] + 0.06, 
        box["label"], 
        ha='center', 
        va='center', 
        fontsize=10
    )

# Ajouter les flèches
arrows = [
    [(0.225, 0.81), (0.225, 0.77), "↓"],
    [(0.225, 0.59), (0.225, 0.53), "↓"],
    [(0.225, 0.39), (0.225, 0.33), "↓"],
    [(0.225, 0.19), (0.225, 0.13), "↓"],
    [(0.35, 0.11), (0.4, 0.11), "→"],
    [(0.65, 0.11), (0.7, 0.11), "→"]
]

for start, end, symbol in arrows:
    arrow = FancyArrowPatch(
        start, end,
        arrowstyle='->,head_width=0.6,head_length=0.8',
        lw=1.5,
        color='black'
    )
    ax.add_patch(arrow)
    plt.text(
        (start[0] + end[0])/2, 
        (start[1] + end[1])/2 - 0.02, 
        symbol, 
        fontsize=14,
        ha='center'
    )

plt.title("Pipeline d'Analyse de Volatilité des Cryptomonnaies", fontsize=16, pad=20)
plt.tight_layout()
plt.savefig('pipeline_volatilite.png', dpi=300, bbox_inches='tight')
plt.show()
```
